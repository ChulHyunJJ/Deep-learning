```
import numpy as np
```

```
class Model:

  def __init__(self, n_net):
    self.n_net = n_net
    self.network = {}
    self.A = {}
    self.Z = {}

  def sigmoid(X):
    return 1/ (1 + np.exp(-X))

# 1 ~ 10사이의 정수를 반환해주는 난수 발생기
  def gen_randint(self, start=1, end=10, seed=0):
    Rand = np.random.RandomState(seed=self.seed)
    return Rand.randint(self.start, self.end)

# n개의 네트워크 생성
  def make_net(self, x):
    # 첫번째 W1 = (n, m) 의 n 은 입력 x의 끝모양에 맞춘다. 
    self.network['W1'] = np.random.randn(x.shape[-1])
    self.network['B1'] = np.random.randn(x.shape[-1])

    for idx in range(self.n_net):
      self.network['W' + str(idx+2)] = np.random.randn(self.network['B' + str(idx+1)].shape[-1], gen_randint(seed=idx+1))
      self.network['B' + str(idx+2)] = np.random.randn(gen_randint(seed=idx+1))

    return self.network

  def forward(self, x, nets):
    self.A['1'] = np.dot(x, nets['W' + str(1)]) + nets['B' + str(1)]
    self.Z['1'] = sigmoid(self.A['1'])
    for i in range(self.n_net-1):
      self.A[str(i+2)] = np.dot(self.A[str(i+1)], nets['W' + str(i+2)]) + nets['B' + str(i+2)]
      self.Z[str(i+2)] = sigmoid(self.A[str(i+2)])

    return self.Z[str(self.n_net)]
```

```
model = Model(5)
```

```
x = np.array([[1,2,3,4], [4,5,6,7], [7,8,9,10], [3,4,5,6]])
x.shape
```

```
print(model.forward(x, nets))
```
